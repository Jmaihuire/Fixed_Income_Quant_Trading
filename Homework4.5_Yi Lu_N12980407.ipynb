```python
import QuantLib as ql
import quandl
import datetime
import pandas as pd
import numpy as np
```

```python
n = range(1,22) # first 21 (most liquid) rolling Eurodollar futures

token = '7Fe845GwTuAehy2o1guA'
nms = ["CHRIS/CME_ED"+str(i) for i in n]
dfs = [quandl.get(nm, authtoken=token) for nm in nms]
```

```python
# put all settle price into one dataframe
settle_price = dfs[0]['Settle']
for i in range(1,21):
    settle1_price = dfs[i]['Settle']
    settle_price = pd.concat([settle_price,settle1_price], axis=1)
settle_price.columns = ['ED'+str(i) for i in n]
settle_price = settle_price.dropna()
settle_price.head(10)
```

```python
settle = pd.DataFrame()
for i in range(21):
    settle['ED'+str(1+i)] = 100- settle_price.iloc[:,i]
settle.head(10)
```

```python
# IMM matrix
imm = pd.DataFrame(index=settle.index, columns=settle.columns)
ql_imm = ql.IMM()

for i in range(len(settle.index)):
    y = settle.index[i].year
    m = settle.index[i].month
    d = settle.index[i].day
    d1 = ql.Date(d,m,y)
    for j in range(len(settle.columns)):
        imm.iloc[i,j] = ql_imm.nextDate(d1)
        d1 = ql_imm.nextDate(d1)

# convert Quantlib Date to Datetime
for i in range(imm.shape[0]):
    for j in range(imm.shape[1]):
        imm.iloc[i,j] = datetime.datetime(imm.iloc[i,j].year(), imm.iloc[i,j].month(), imm.iloc[i,j].dayOfMonth())

imm.head(10)
```

```python
# calculate actual/constant time to maturity, interpolated yields
actual = pd.DataFrame(columns = imm.columns, index = imm.index) # actual time to maturity
constant = actual  # constant time to maturity
yields = actual # linearly interpolated yields
```

```python
# actual time to maturity
actual = pd.DataFrame(columns = imm.columns, index = imm.index)

for i in range(actual.shape[0]):
    for j in range(actual.shape[1]):
        actual.iloc[i,j] = (imm.iloc[i,j] - imm.index[i]).days

actual.head()
```

```python
# constant time to maturity
constant = pd.DataFrame(columns = imm.columns, index = imm.index)

for i in range(constant.shape[0]):
    for j in range(constant.shape[1]):
        y = constant.index[i].year
        m = constant.index[i].month
        d = constant.index[i].day
        y_add = (m+3*(j+1))/12
        m_new = (m+3*(j+1))%12
        if m_new == 0:
            y_add = y_add -1
            m_new = 12
        y_add = int(y_add // 1) # get the int before decimal
        try: 
            constant.iloc[i,j] =datetime.datetime(y+y_add, m_new, d) - constant.index[i]
        except ValueError:
            try:
                constant.iloc[i,j] = datetime.datetime(y+y_add, m_new, d-1) - constant.index[i]
            except ValueError:
                try:
                    constant.iloc[i,j] = datetime.datetime(y+y_add, m_new, d-2) - constant.index[i]
                except ValueError:
                    try:
                        constant.iloc[i,j] = datetime.datetime(y+y_add, m_new, d-3) - constant.index[i]
                    except ValueError:
                        constant.iloc[i,j] = datetime.datetime(y+y_add, m_new, d-4) - constant.index[i]  #2/27/2004

constant = constant.applymap(lambda x: x.days)
constant.head()
```

```python
# linearly interpolated yields
yields = pd.DataFrame(columns = imm.columns, index = imm.index)

for i in range(yields.shape[0]):
    for j in range(yields.shape[1]): # perform linear interpolation
        yields.iloc[i,j] = np.interp(constant.iloc[i,j], list(actual.iloc[i,:]), list(settle.iloc[i,:]))

yields.head()
```
