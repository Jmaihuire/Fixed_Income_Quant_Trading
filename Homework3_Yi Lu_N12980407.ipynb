```python
import quandl
import numpy as np
import pandas as pd
```

```python
n = range(1,9) # first 8 (most liquid) rolling Eurodollar futures
nms = ["CHRIS/CME_ED"+str(i) for i in n]
dfs = [quandl.get(nm) for nm in nms]
```

```python
# put all settle price into one dataframe
settle_price = dfs[0]['Settle']
for i in range(1,8):
    settle1_price = dfs[i]['Settle']
    settle_price = pd.concat([settle_price,settle1_price], axis=1)
settle_price.columns = ['ED'+str(i) for i in n]
settle_price
```

### 1. Using Sample1 construct all possible cointegrated WSPR using CCA and/or level regression.

```python
settle = pd.DataFrame()
for i in range(8):
    settle['ED'+str(1+i)] = 100- settle_price.iloc[:,i]
settle.index = pd.to_datetime(settle.index)
settle_sample1 = settle['2004-01-01':'2006-12-31']
settle_sample1 = settle_sample1.dropna()
settle_sample1.head()
```

```python
import statsmodels.api as sm

def CCA_Chou_Ng(data_set):
    
    #data_set is pandas dataframe
    df_lag = data_set.shift(1).dropna()
    df = data_set.drop(data_set.index[0]).dropna()
    n = len(data_set.columns)
    
    #X(t) ~ M_1 + X(t-1)
    X = df_lag.as_matrix()
    X_I = sm.add_constant(X)
    Y = df.as_matrix()
    l1 = sm.OLS(Y,X_I).fit()
    B=l1.params[1:(n+1)]
    
    #X(t-1) ~ M_2 + X(t)
    Y_I = sm.add_constant(Y)
    l2 = sm.OLS(X,Y_I).fit()
    A=l2.params[1:(n+1)]
    C = np.dot(A,B)
    eig_val, eig_vec = np.linalg.eig(C)
    ascor = np.argsort(eig_val)
    eig_val, eig_vec = eig_val[ascor], eig_vec[:, ascor]
    return (eig_vec[1,0]/eig_vec[0,0])
```

```python
def level_regression(data_set):
    data_set = data_set.dropna()
    
    X = data_set.iloc[:,1]
    X_I = sm.add_constant(X)
    Y = data_set.iloc[:,0]
    l3 = sm.OLS(Y,X_I).fit()
    return l3.params[1]
```

```python
from itertools import combinations

combs = list(combinations(range(8),2))
print ('There are',len(combs),'combinations')
print ('==========================')

weights_levelreg = np.ones((len(combs),2))
weights_CCA = np.ones((len(combs),2))

for i in range(len(combs)):
    test = settle_sample1.iloc[:,list(combs[i])]
    
    weights_levelreg[i,1] = -level_regression(test)
    weights_CCA[i,1] = CCA_Chou_Ng(test)

weights_levelreg_matrix = pd.DataFrame(weights_levelreg, 
                                       index=['ED'+str(i[0])+'~'+'ED'+str(i[1]) for i in combinations(range(1,9),2)])
weights_levelreg_matrix.columns = ['1st weight', '2nd weight']
weights_CCA_matrix = pd.DataFrame(weights_CCA, 
                                       index=['ED'+str(i[0])+'~'+'ED'+str(i[1]) for i in combinations(range(1,9),2)])
weights_CCA_matrix.columns = ['1st weight', '2nd weight']

print ('Weights under Level Regression:')
print (weights_levelreg_matrix)
print ('==========================')
print ('Weights under CCA:')
print (weights_CCA_matrix)
```

### 2. Inspect stationarity of all WSPRs in Sample1, and choose the best 5

```python
# stationary test
import statsmodels.tsa.stattools as ts

def stationary_test(data_set, weight):
    data = (weight * data_set).sum(axis=1)
    adf = ts.adfuller(data, maxlag=1)
    return adf[1] # p-value
```

```python
pvalue_levelreg = []
pvalue_CCA = []

for i in range(len(combs)):
    test = settle_sample1.iloc[:,list(combs[i])]
    test = test.dropna()
    
    weight_levelreg = list(weights_levelreg_matrix.iloc[i,:])
    weight_CCA = list(weights_CCA_matrix.iloc[i,:])
    
    pvalue_levelreg.append(stationary_test(test, weight_levelreg))
    pvalue_CCA.append(stationary_test(test, weight_CCA))

pvalue_levelreg_matrix = pd.DataFrame(pvalue_levelreg, 
                                       index=['ED'+str(i[0])+'~'+'ED'+str(i[1]) for i in combinations(range(1,9),2)])
pvalue_levelreg_matrix.columns = ['p-value']
pvalue_levelreg_matrix['Stationary'] = pvalue_levelreg_matrix['p-value']<0.05
pvalue_CCA_matrix = pd.DataFrame(pvalue_CCA, 
                                       index=['ED'+str(i[0])+'~'+'ED'+str(i[1]) for i in combinations(range(1,9),2)])
pvalue_CCA_matrix.columns = ['p-value']
pvalue_CCA_matrix['Stationary'] = pvalue_CCA_matrix['p-value']<0.05

print ('Stationary Test of Level Regression:')
print (pvalue_levelreg_matrix)
print ('====================================')
print (' ')
print ('Stationary Test of CCA:')
print (pvalue_CCA_matrix)
```

```python
# choose best 5
def best_5(pvalue_matrix, weights_matrix):
    sorted_matrix = pvalue_matrix['p-value'].sort_values()
    outcome = weights_matrix[pvalue_matrix['p-value']
                             <sorted_matrix[5]].join(pvalue_matrix[pvalue_matrix['p-value']<sorted_matrix[5]])
    outcome = outcome.sort_values(by=['p-value'])
    return outcome
```

```python
best5_levelreg = best_5(pvalue_levelreg_matrix, weights_levelreg_matrix)
best5_CCA = best_5(pvalue_CCA_matrix, weights_CCA_matrix)

print ('The Best 5 Stationary in Level Regression:')
print (best5_levelreg)
print ('==========================================')
print ('The Best 5 Stationary in CCA:')
print (best5_CCA)
```

### 3. Inspect stationarity of all WSPR 3m out-of-sample, and choose the best 5

```python
# 3 month out-of-sample
settle_sample2 = settle['2007-01-01':'2007-03-31']
settle_sample2 = settle_sample2.dropna()
```

```python
pvalue_levelreg2 = []
pvalue_CCA2 = []

for i in range(len(combs)):
    test2 = settle_sample2.iloc[:,list(combs[i])]
    test2 = test2.dropna()
    
    weight_levelreg = list(weights_levelreg_matrix.iloc[i,:])
    weight_CCA = list(weights_CCA_matrix.iloc[i,:])
    
    pvalue_levelreg2.append(stationary_test(test2, weight_levelreg))
    pvalue_CCA2.append(stationary_test(test2, weight_CCA))

pvalue_levelreg_matrix2 = pd.DataFrame(pvalue_levelreg2, 
                                       index=['ED'+str(i[0])+'~'+'ED'+str(i[1]) for i in combinations(range(1,9),2)])
pvalue_levelreg_matrix2.columns = ['p-value']
pvalue_levelreg_matrix2['Stationary'] = pvalue_levelreg_matrix2['p-value']<0.05
pvalue_CCA_matrix2 = pd.DataFrame(pvalue_CCA2, 
                                       index=['ED'+str(i[0])+'~'+'ED'+str(i[1]) for i in combinations(range(1,9),2)])
pvalue_CCA_matrix2.columns = ['p-value']
pvalue_CCA_matrix2['Stationary'] = pvalue_CCA_matrix2['p-value']<0.05

print ('Stationary Test of Level Regression (out-of-sample):')
print (pvalue_levelreg_matrix2)
print ('=======================================================')
print (' ')
print ('Stationary Test of CCA (out-of-sample):')
print (pvalue_CCA_matrix2)
```

```python
best5_levelreg2 = best_5(pvalue_levelreg_matrix2, weights_levelreg_matrix)
best5_CCA2 = best_5(pvalue_CCA_matrix2, weights_CCA_matrix)

print ('The Best 5 Stationary in Level Regression:')
print (best5_levelreg2)
print ('==========================================')
print ('The Best 5 Stationary in CCA:')
print (best5_CCA2)
```
